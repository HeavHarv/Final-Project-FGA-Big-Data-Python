# Meng-import data ke database


```
#Membuat method untuk memasukkan data external ke dalam database

def pushCSVdatasetToDB(Tahun,Populasi,Kadar_CO2,Kadar_CH4,Kadar_N2O,Suhu_Bumi):
#memasukkan niali variabel ke dalam tabel
    db = sqlite3.connect("fga_big_data.db")    
    cmd = "insert into climate(Tahun,Populasi_Triliun,Kadar_CO2_ppm,Kadar_CH4_ppb,Kadar_N20_ppb,Suhu_Bumi_°C) values('{}','{}','{}','{}','{}','{}')".format(Tahun,Populasi,Kadar_CO2,Kadar_CH4,Kadar_N2O,Suhu_Bumi)
    db.execute(cmd)
    db.commit()
```

```
data = pd.read_csv("climate_data.csv")
data.shape
data.head()
```

```
#memasukkan 30 baris data dari climate.csv ke fga_big_data.db
for i in range(0,30):
  pushCSVdatasetToDB(data.iloc[i][0],data.iloc[i][1],data.iloc[i][2],data.iloc[i][3],data.iloc[i][4],data.iloc[i][5])
```
